```python
# Найдем промежутки возрастания и убывания функции.
# промежутки будем искать через производную 
expr_diff = expr.diff(x)
print(expr_diff)
```

```python
#Найдем экстремумы производной и определим промежутку возрастания и убывания функции.
```

```python
#найдем корни уравнения производной с применением библиотеки numpy
root_diff_by_numpy = np.roots([-48, -54, 10, 10])
root_diff_by_numpy
```

```python
#найдем корни уравнения производной с применением библиотеки sympy
root_diff_by_sympy = sympy.solvers.solve(expr_diff, x)
root_diff_by_sympy
```

```python
root_diff_by_sympy[0]
```

```python
# странно, что корни первой производной, найденные с применением библиотеки numpy действительные, а корни,
# определенные с применением библиотекеи sympy имеют мнимую часть.
```

```python
# проверим значения экстемумов, определенных с применением библиотеки numpy
# уравнение производной -48*x**3 - 54*x**2 + 10*x + 10 = 0
for el in root_diff_by_numpy:
    print(-48*el**3 - 54*el**2 + 10*el + 10)
```

```python
# Получаем два ненулевых значения. Ненулевые значения объясняются погрешностью вычисления чисел с плавающей точкой
```

```python
# проверим значения экстемумов, определенных с применением библиотеки sympy
for el in root_diff_by_sympy:
    print(expr_diff.subs(x, el).evalf())
```

```python
# видим, что значение не равно ноль, это погрешность вычисления числа с плавающей точкой.
# применим вычисление с отсечением погрешности
for el in root_diff_by_sympy:
    print(expr_diff.subs(x, el).evalf(chop=True))
```

```python
# найдем значения второй производной в точках экстемума функции
# т.к. полученыне с помощью sympy значения второй производной функции корней первой производной комплексные, то 
val_diff2 = []
for i, el in enumerate(root_diff_by_sympy):
    val_diff2.append(expr_diff2.subs(x ,el).evalf())
    print(f'expr({el}) = {val_diff2} {"maximun" if val_diff2[i].is_positive > 0 else "minimum"} ')
```

```python
# парсинг многочлена с использованием регулярных выражений
import re
```

```python
pattern = r'([-+][^=]?\d*\.?\d*)(x\^*\d*)*'   # добавлена возможность выделения коэффициентов записанных в виде десятичной дроби
#input_str = '-22x^3+34x^2+5x+18=0'
#input_str = '-2.2x^3+5x+18=0'  
input_str = '+15-2.2x^3+5x=0'                 # расположение членов многочлена не имеет значения
res_as_list = re.findall(pattern, input_str)
print(res_as_list)
res_as_list = [(el[0], 0 if len(el[1]) == 0 else (int(el[1].replace('^', '').replace('x', '')) if '^' in el[1] else 1)) for el in res_as_list]
print(res_as_list)
max_n = max([el[1] for el in res_as_list])
print(max_n)
r = []
for i in range(max_n + 1):
    for el in res_as_list:
        if el[1] == i:
            if '.' in el[0]:
                r.insert(0, float(el[0])) 
            else:
                r.insert(0, int(el[0]))      # если нет десятичной точки, то конвертация в int для повышения точности    
            break
    else:
        r.insert(0, 0)
print(r)                                     # строка коэффициентов
```
